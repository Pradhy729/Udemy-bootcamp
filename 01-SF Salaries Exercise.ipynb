** How many Job Titles were represented by only one person in 2013? (e.g. Job Titles with only one occurence in 2013?) **

```python
just2013_df = df[df['Year'] == 2013]
num_unique_2013_series = just2013_df.JobTitle.value_counts()
num_unique_2013_series[num_unique_2013_series == 1].size
```

** How many people have the word Chief in their job title? (This is pretty tricky) **

```python
chief = lambda x: 'chief' in x.lower()
df[df.JobTitle.apply(chief)]['EmployeeName'].nunique()
```

** Bonus: Is there a correlation between length of the Job Title string and Salary? **

```python
series_of_JT_lengths = df['JobTitle'].apply(lambda x: len(x))
newdf = pd.concat([df,series_of_JT_lengths.rename('lengthOfString')],axis=1,)
newdf['lengthOfString'].corr(newdf['TotalPayBenefits'])
```
